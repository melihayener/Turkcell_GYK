Yapılması Gerekenler:
En çok saat harcayan çalışanı bulun.
Tamamlanan proje sayısı en yüksek olan çalışanı belirleyin.
Her departmandaki toplam çalışma saatlerini hesaplayın.
En çok proje yapan departmanı bulun.
Tüm projelerin tamamlanma yüzdesini hesaplayın.

```python
employees = [
    {"name": "Ahmet", "department": "IT", "projects": [{"name": "CRM", "completed": True, "hours": 120},
    {"name": "Security", "completed": False, "hours": 80}]},
    
    {"name": "Elif", "department": "HR", "projects": [{"name": "Hiring System", "completed": True, "hours": 90}, 
    {"name": "Employee Portal", "completed": True, "hours": 100}]},
   
    {"name": "Mehmet", "department": "IT","projects": [{"name": "Cloud Migration", "completed": True, "hours": 150},
    {"name": "Data Analysis", "completed": True, "hours": 130}]},
    
    {"name": "Ayşe", "department": "Marketing","projects": [{"name": "Social Media", "completed": True, "hours": 60},
    {"name": "SEO", "completed": False, "hours": 50}]}
]

employee_hours = {}
complete_project = {}
department_hours = {}
department_project = {}
total_projects = 0
complete_projects = 0

for item in employees:
    for detail in item["projects"]:
        if item["name"] in employee_hours:
            employee_hours[item["name"]] += detail["hours"]
        else:
            employee_hours[item["name"]] = detail["hours"]

        total_projects += 1
        
        if detail["completed"] == True:
            complete_projects += 1  
            
            if item["name"] in complete_project:
                complete_project[item["name"]] += 1
            else:
                complete_project[item["name"]] = 1
            
            if item["department"] in department_project:
                department_project[item["department"]] += 1
            else:
                department_project[item["department"]] = 1
        
        if item["department"] in department_hours:
            department_hours[item["department"]] += detail["hours"]
        else:
            department_hours[item["department"]] = detail["hours"]
            
#1 En çok saat harcayan çalışanı bulun.
employee_hours = sorted(employee_hours.items(), key = lambda x: x[1], reverse=True)
print("En çok saat harcayan çalışan",employee_hours[0][0])  

#2 Tamamlanan proje sayısı en yüksek olan çalışanı belirleyin
print(complete_project)
complete_project = sorted(complete_project.items(), key = lambda x: x[1], reverse=True)
print("Tamamlanan proje sayısı en yüksek olan çalışan",complete_project[0][0])

#3 Her departmandaki toplam çalışma saatlerini hesaplayın.
print(department_hours)
department_hours = sorted(department_hours.items(), key = lambda x: x[1], reverse=True)
print("Her departmandaki toplam çalışma saatleri",department_hours[0][0])

#4 En çok proje yapan departmanı bulun.
print(department_project)
department_project = sorted(department_project.items(), key = lambda x: x[1], reverse=True)
print("En çok proje yapan departman",department_project[0][0])

#5 Tüm projelerin tamamlanma yüzdesini hesaplayın.
print("Tüm projelerin tamamlanma yüzdesi","%",100*(complete_projects/total_projects))
```

Yapılması Gerekenler:
1_Stokta toplam kaç ürün olduğunu hesaplayın.
2_En değerli ürünü (price * quantity en yüksek olan) bulun.
3_Her kategori için toplam stok değerini hesaplayan bir sözlük oluşturun.
4_Stok miktarı en düşük olan ürünü bulun.
5_Ortalama ürün fiyatını hesaplayın.

```python
stock = [
    {"category": "Meyve", "items": [{"name": "Elma", "price": 10, "quantity": 50}, 
                                    {"name": "Muz", "price": 15, "quantity": 30}]},
    
    {"category": "Sebze", "items": [{"name": "Domates", "price": 8, "quantity": 60}, 
                                    {"name": "Patates", "price": 5, "quantity": 100}]},
    
    {"category": "İçecek", "items": [{"name": "Maden Suyu", "price": 3, "quantity": 150}, 
                                     {"name": "Kola", "price": 12, "quantity": 80}]}
]

total_item = 0
items_lux = {}
category_items = {}
items_stock = {}
total_price = 0

for item in stock:
    for key in item["items"]:
        total_price += key["price"]
        total_item += key["quantity"]
        items_lux[key["name"]] = (key["price"] * key["quantity"])
        
        if item["category"] in category_items:
            category_items[item["category"]] += key["quantity"]
        else:
            category_items[item["category"]] = key["quantity"]
        
        if key["name"] in items_stock:
            items_stock[key["name"]] += key["quantity"]
        else:
            items_stock[key["name"]] = key["quantity"]
            
#1_Stokta toplam kaç ürün olduğunu hesaplayın.            
print("Stokta olan toplam ürün sayısı:", total_item)   

#2_En değerli ürünü (price * quantity en yüksek olan) bulun.
items_lux = sorted(items_lux.items(), key = lambda x:x[1], reverse=True)
print(items_lux[0][0])

#3_Her kategori için toplam stok değerini hesaplayan bir sözlük oluşturun.
print("Her kategori için toplam stok değerini hesaplayan sözlük:",category_items)

#4_Stok miktarı en düşük olan ürünü bulun.
items_stock = sorted(items_stock.items(), key = lambda x:x[1], reverse=False)
print("Stok miktarı en düşük olan ürün:", items_stock[0][0])

#5_Ortalama ürün fiyatını hesaplayın.
print("Ortalama ürün fiyatı",total_price/len(items_stock))
```

4_Tüm öğrencilerin genel ortalamasını hesaplayın.
5_Her ders için en yüksek notu alan öğrenciyi bulun.

```python
students = [
    {"name": "Ali", "grades": {"math": 85, "physics": 90, "history": 78}},
    {"name": "Ayşe", "grades": {"math": 92, "physics": 88, "history": 95}},
    {"name": "Mehmet", "grades": {"math": 75, "physics": 80, "history": 85}},
    {"name": "Zeynep", "grades": {"math": 90, "physics": 85, "history": 88}}
]

students_avg = {}
lessons_avg = {}
sum_grade = 0
sum_lesson = 0

for item in students:
    for key,value in item["grades"].items():
        sum_grade += value
        sum_lesson +=1
        if item["name"] in students_avg:
            students_avg[item["name"]] += value
        else:
            students_avg[item["name"]] = value
        
        if key in lessons_avg:
            lessons_avg[key] += value
        else:
            lessons_avg[key] = value

students_avg = sorted(students_avg.items(), key = lambda x:x[1], reverse=True)

#1_En yüksek ortalama notuna sahip öğrenciyi bulun.
print("En yüksek ortalama notuna sahip öğrenci",students_avg[0][0])

#2_Her dersin ortalama notunu hesaplayın.
for key,value in lessons_avg.items():
    value = value/len(students)
    lessons_avg[key] = value
print(lessons_avg)

#3_Ortalama notu 85’in üzerinde olan öğrencileri listeleyin.
students_ = []
avg={}
for item in students_avg:
    avg[item[0]] = item[1]/3   
    if avg[item[0]] > 85:
        students_.append(item[0])
print("Ortalama notu 85’in üzerinde olan öğrenciler",students_) 

#4_Tüm öğrencilerin genel ortalamasını hesaplayın.
print("Tüm öğrencilerin genel ortalaması:",(sum_grade/sum_lesson))  

#5_Her ders için en yüksek notu alan öğrenciyi bulun.
math = {}
physics={}
history={}
for student in students:
    math[student["name"]] = student["grades"]["math"]
    physics[student["name"]] = student["grades"]["physics"]
    history[student["name"]] = student["grades"]["history"]
    
math = sorted(math.items(), key = lambda x: x[1], reverse=True)
physics = sorted(physics.items(), key = lambda x: x[1], reverse=True)
history = sorted(history.items(), key = lambda x: x[1], reverse=True)

print("Matematik:",math[0][0])
print("Fizik:",physics[0][0])
print("Tarih:",history[0][0])
```

]
Yapılması gerekenler:
1_En yüksek matematik notuna sahip öğrencinin adını bulun.
2_Tüm derslerin ortalama notlarını içeren bir sözlük oluşturun.
3_Ortalama notu 85'in üzerinde olan öğrencileri bir liste halinde döndürün.
Bunu en verimli şekilde Python kodu ile nasıl yaparsınız?

```python
students = [
    {"name": "Ali", "grades": {"math": 85, "physics": 90, "history": 78}},
    {"name": "Ayşe", "grades": {"math": 92, "physics": 88, "history": 95}},
    {"name": "Mehmet", "grades": {"math": 75, "physics": 80, "history": 85}},
    {"name": "Zeynep", "grades": {"math": 90, "physics": 85, "history": 88}},
]
   
#1_En yüksek matematik notuna sahip öğrencinin adını bulun. (1. cevap)
grades_math = []
for item in students:
    grades_math.append(item["grades"]["math"])

max_grade = max(grades_math)

for item in students:
    if item["grades"]["math"] == max_grade:
        print("En yüksek matematik notuna sahip öğrencinin adı:",item["name"])

#2_Tüm derslerin ortalama notlarını içeren bir sözlük oluşturun.
math = 0
physics= 0
history = 0

for item in students:
    math += item["grades"]["math"]
    physics += item["grades"]["physics"]
    history += item["grades"]["history"]
    
lessons = {}
lessons["math"] = math/len(students)
lessons["physics"] = physics/len(students)
lessons["history"] = history/len(students)

print("Tüm derslerin ortalama notlarını içeren sözlük:",lessons)

#3_Ortalama notu 85'in üzerinde olan öğrencileri bir liste halinde döndürün.
students_avg = []
for item in students:
    avg  = (item["grades"]["math"]+
                     item["grades"]["physics"]+
                     item["grades"]["history"]) / 3
    if avg > 85:
        students_avg.append(item["name"])
print("Ortalama notu 85'in üzerinde olan öğrencilerin listesi:",students_avg)
```

```python
employees = [
    {
        "name": "Ahmet",
        "department": "IT",
        "salary": 8500,
        "tasks": [
            {"task": "Database Optimization", "completed": True, "effort": 5},
            {"task": "API Development", "completed": True, "effort": 8},
            {"task": "Security Audit", "completed": False, "effort": 6},
        ],
    },
    {
        "name": "Elif",
        "department": "HR",
        "salary": 7500,
        "tasks": [
            {"task": "Recruitment", "completed": True, "effort": 7},
            {"task": "Training", "completed": True, "effort": 4},
            {"task": "Policy Update", "completed": False, "effort": 3},
        ],
    },
    {
        "name": "Mehmet",
        "department": "IT",
        "salary": 9200,
        "tasks": [
            {"task": "Cloud Migration", "completed": True, "effort": 9},
            {"task": "Load Balancing", "completed": True, "effort": 7},
            {"task": "System Monitoring", "completed": True, "effort": 5},
        ],
    },
    {
        "name": "Ayşe",
        "department": "Marketing",
        "salary": 6800,
        "tasks": [
            {"task": "Social Media Campaign", "completed": True, "effort": 6},
            {"task": "SEO Optimization", "completed": False, "effort": 4},
            {"task": "Brand Strategy", "completed": True, "effort": 7},
        ],
    },
]

#1_En yüksek maaşa sahip çalışanı bulun.
salary = []
for item in employees:
    salary.append(item["salary"])

max_salary = max(salary)

for item in employees:
    if item["salary"] == max_salary:
        print("En yüksek maaşa sahip çalışan:",item["name"])

#2_Her departmandaki çalışanların toplam maaşlarını içeren bir sözlük oluşturun.
departman_maas = {}
for item in employees:
    if item["department"] in departman_maas:
        departman_maas[item["department"]] += item["salary"]
    else:
        departman_maas[item["department"]] = item["salary"]
        
print("Her departmandaki çalışanların toplam maaşlarını içeren sözlük",departman_maas)

#3_Tüm görevlerin tamamlanma yüzdesini hesaplayın. (Tamamlanan görev sayısı / Toplam görev sayısı)
total_task = 0
completed_task = 0
employees_effort = {}
full_task = set()
enaz1_task = set()

for item in employees:
    full_task.add(item["name"])
    for tasks in item["tasks"]:
        total_task += 1
        if tasks["completed"] :
            completed_task +=1
            enaz1_task.add(item["name"])
            if item["name"] in employees_effort:
                employees_effort[item["name"]] += tasks["effort"]
            else:
                employees_effort[item["name"]] = tasks["effort"]
        else:
            full_task.discard(item["name"])
            

print("Tüm görevlerin tamamlanma yüzdesi:","%",100*(completed_task/total_task))

#4_Çalışanları tamamladıkları görevlerin toplam "effort" değerine göre sıralayarak listeleyin.
efforts = sorted(employees_effort.items(), key=lambda x : x[1], reverse=True)
print("Çalışanları tamamladıkları görevlerin toplam effort değerine göre sıralandığı liste:",efforts)

#5_Görevlerini %100 tamamlamış çalışanları listeleyin.
print("Görevlerini %100 tamamlamış çalışanlar:",full_task)

#6_Her departman için, en az bir görevini tamamlamış çalışanların adlarını içeren bir sözlük oluşturun.
print("Her departman için, en az bir görevini tamamlamış çalışanların adlarını içeren sözlük:",enaz1_task)
```

Yapılması Gerekenler:
1_Toplam sipariş tutarı en yüksek olan müşteriyi bulun.
2_Her müşterinin toplam sipariş tutarını içeren bir sözlük oluşturun.
3_En çok satılan ürünü bulun. (Birden fazla müşteri tarafından sipariş edilen ürün)
4_Ortalama sipariş tutarını hesaplayın.
5_Siparişlerinde toplam tutarı 10.000 TL’nin üzerinde olan müşterileri listeleyin.

```python
orders = [
    {"customer": "Ali", "items": [{"name": "Laptop", "price": 15000}, 
                                  {"name": "Mouse", "price": 500}]},
    {"customer": "Ayşe", "items": [{"name": "Phone", "price": 12000}, 
                                   {"name": "Headphones", "price": 2000}]},
    {"customer": "Mehmet", "items": [{"name": "Tablet", "price": 7000}, 
                                     {"name": "Keyboard", "price": 1000}, 
                                     {"name": "Mouse", "price": 600}]},
    {"customer": "Zeynep", "items": [{"name": "Monitor", "price": 8000}, 
                                     {"name": "Phone", "price": 11000}]},
]

#1_Toplam sipariş tutarı en yüksek olan müşteriyi bulun.
total_order = {}
max_items = {}
for item in orders:
    for i in item["items"]:
        if i["name"] in max_items:
            max_items[i["name"]] += 1
        else:
            max_items[i["name"]] = 1
            
        if item["customer"] in total_order:
            total_order[item["customer"]] += i["price"]
        else:
            total_order[item["customer"]] = i["price"]
            
max_customer = total_order.copy()
max_customer = sorted(max_customer.items(), key = lambda x : x[1], reverse=True)
print("Toplam sipariş tutarı en yüksek olan müşteri",max_customer[0][0])

#2_Her müşterinin toplam sipariş tutarını içeren bir sözlük oluşturun.
print("Her müşterinin toplam sipariş tutarını içeren sözlük",total_order)

#3_En çok satılan ürünü bulun. (Birden fazla müşteri tarafından sipariş edilen ürün)
max_items = sorted(max_items.items(), key = lambda x:x[1], reverse=True)
print("En çok satılan ürün:", max_items[0][0])

#4_Ortalama sipariş tutarını hesaplayın.
avg = 0
for key,value in total_order.items():
    avg += value    
print("Ortalama sipariş tutarı:",avg/len(total_order))

#5_Siparişlerinde toplam tutarı 10.000 TL’nin üzerinde olan müşterileri listeleyin.
order_max = []
for customer in total_order:
    if total_order[customer] > 10000:
        order_max.append(total_order[customer])
print("Siparişlerinde toplam tutarı 10.000 TL’nin üzerinde olan müşteriler", order_max)
```
